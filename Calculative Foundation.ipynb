# PART A: VECTOR OPERATIONS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eig, det, inv
```

```python
df = pd.read_csv("C:\\Users\\tanaa\\Desktop\\student_marks_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.drop("Student_ID", axis=1).values
print("Matrix:\n", X)
```

```python
v1 = X[0]
v2 = X[1]

print("Vector 1:", v1)
print("Vector 2:", v2)
```

```python
print("Norm-1:", np.linalg.norm(v1, ord=1))
print("Norm-2:", np.linalg.norm(v1))
```

```python
dot = np.dot(v1, v2)
print("Dot Product:", dot)
```

```python
angle = np.arccos(dot / (np.linalg.norm(v1) * np.linalg.norm(v2)))
print("Angle (radians):", angle)
```

```python
cross = np.cross(v1[:3], v2[:3])
print("Cross Product:", cross)
```

```python
proj = (np.dot(v1, v2) / np.dot(v2, v2)) * v2
print("Projection of v1 on v2:", proj)
```

# PART B: MATRIX OPERATIONS

```python
XT = X.T
print("Transpose:\n", XT)
```

```python
mul = np.dot(X, XT)
print("Matrix Multiplication:\n", mul)
```

```python
cov = np.cov(XT)
print("Covariance Matrix:\n", cov)
```

```python
print("Determinant:", det(cov))
```

```python
print("Determinant:", det(cov))
```

```python
try:
    inv_matrix = inv(cov)
    print("Inverse Matrix:\n", inv_matrix)
except:
    print("Matrix not invertible")
```

# PART C: VISUALIZATION

# PART D: EIGENVALUES & EIGENVECTORS

```python
eigenvalues, eigenvectors = eig(cov)

print("Eigenvalues:\n", eigenvalues)
print("Eigenvectors:\n", eigenvectors)
```

# PART E: PCA (DIMENSION REDUCTION)

```python
# Sort eigenvalues
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Reduce to 2D
X_reduced = np.dot(X, eigenvectors[:, :2])

print("Reduced Data:\n", X_reduced)
```
